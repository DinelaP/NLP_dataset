```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver

# Kreiraj WebDriver
driver = web_driver()

# Statiƒçki poƒçetni URL
static_url = "https://www.sum.ba/"
driver.get(static_url)

# Provjera uƒçitavanja stranice
print("Uƒçitana stranica:", driver.title)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import fitz  # PyMuPDF
import requests
import os
import time

putanja_fajla = "/content/drive/MyDrive/PDF/dokumenti4_pdf.txt"

url = "https://www.sum.ba/sum/dokumenti/propisi-u-bih"
driver.get(url)
time.sleep(3)

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
driver.quit()

pdf_links = soup.find_all("a", href=True)
pdf_links = [a for a in pdf_links if a["href"].endswith(".pdf")]

with open(putanja_fajla, "w", encoding="utf-8") as f:
    stranica = 1
    for a in pdf_links:
        naslov = a.get_text(strip=True)
        href = a["href"]
        if not href.startswith("http"):
            href = "https://web-admin.sum.ba" + (href if href.startswith("/") else "/" + href)

        print(f"üîó Preuzimam: {naslov}")

        try:
            pdf_response = requests.get(href)
            pdf_response.raise_for_status()

            temp_pdf = f"temp_{stranica}.pdf"
            with open(temp_pdf, "wb") as temp_file:
                temp_file.write(pdf_response.content)

            with fitz.open(temp_pdf) as doc:
                tekst = ""
                for page in doc:
                    tekst += page.get_text()

            if not tekst.strip():
                tekst = "Tekst nije dostupan."

            os.remove(temp_pdf)

        except Exception as e:
            print(f"‚ö†Ô∏è Gre≈°ka kod PDF-a: {e}")
            tekst = "N/A"

        f.write(f"Novina: Sveuƒçili≈°te u Mostaru - SUM\n")
        f.write(f"Datum: N/A\n")
        f.write(f"Kategorija: Meƒëunarodni propisi\n")
        f.write(f"Nadnaslov: N/A\n")
        f.write(f"Naslov: {naslov}\n")
        f.write(f"Podnaslov: N/A\n")
        f.write(f"Strana: {stranica}\n")
        f.write(f"Autor(i): N/A\n")
        f.write("Tekst:\n")
        f.write(f"{tekst}\n")
        f.write("=" * 80 + "\n")

        stranica += 1
        time.sleep(1)

print(f"\n‚úÖ Zavr≈°eno. Podaci sa PDF-ova su saƒçuvani u '{putanja_fajla}'.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import fitz  # PyMuPDF
import requests
import os
import time

# Putanja za ƒçuvanje podataka
putanja_fajla = "/content/drive/MyDrive/PDF/pravilnici_pdf.txt"


wait = WebDriverWait(driver, 10)

# Otvori poƒçetnu stranicu
driver.get("https://www.sum.ba/objave/pravilnici")
time.sleep(3)


broj_stranica = 12
stranica = 1

with open(putanja_fajla, "w", encoding="utf-8") as f:
    for page_num in range(1, broj_stranica + 1):
        print(f"üìÑ Obrada stranice {page_num}")
        time.sleep(2)

        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")


        a_tagovi = soup.find_all("a", href=True, string=lambda t: t and "ovdje" in t.lower())

        for a_tag in a_tagovi:
            href = a_tag["href"]
            if not href.endswith(".pdf"):
                continue


            parent_p = a_tag.find_parent("p")
            naslov = parent_p.get_text(strip=True).replace("\n", " ") if parent_p else "Pravilnik"

            if not href.startswith("http"):
                href = "https://web-admin.sum.ba" + href

            print(f"üîó Preuzimam: {naslov}")
            try:
                pdf_response = requests.get(href)
                pdf_response.raise_for_status()

                temp_pdf = f"temp_{stranica}.pdf"
                with open(temp_pdf, "wb") as temp_file:
                    temp_file.write(pdf_response.content)

                with fitz.open(temp_pdf) as doc:
                    tekst = ""
                    for page in doc:
                        tekst += page.get_text()

                os.remove(temp_pdf)

                if not tekst.strip():
                    tekst = "Tekst nije dostupan."

            except Exception as e:
                print(f"‚ö†Ô∏è Gre≈°ka kod PDF-a: {e}")
                tekst = "N/A"

            f.write(f"Novina: Sveuƒçili≈°te u Mostaru - SUM\n")
            f.write(f"Datum: N/A\n")
            f.write(f"Kategorija: Pravilnici\n")
            f.write(f"Nadnaslov: N/A\n")
            f.write(f"Naslov: {naslov}\n")
            f.write(f"Podnaslov: N/A\n")
            f.write(f"Strana: {stranica}\n")
            f.write(f"Autor(i): N/A\n")
            f.write("Tekst:\n")
            f.write(f"{tekst}\n")
            f.write("=" * 80 + "\n")

            stranica += 1
            time.sleep(1)


        if page_num < broj_stranica:
            try:
                dugmad = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "button.v-pagination__item")))
                for dugme in dugmad:
                    if dugme.text == str(page_num + 1):
                        driver.execute_script("arguments[0].click();", dugme)
                        break
            except Exception as e:
                print(f"‚ö†Ô∏è Gre≈°ka kod klika na paginaciju: {e}")
                break

driver.quit()
print(f"\n‚úÖ Zavr≈°eno. Podaci sa PDF-ova su saƒçuvani u '{putanja_fajla}'.")
```
